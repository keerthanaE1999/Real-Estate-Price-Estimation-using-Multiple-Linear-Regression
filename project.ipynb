###**UNIVARIENT ANALYSIS**

```python
df = df.drop(columns=["No"])
```

```python
df.columns = [
    "transaction_date", "house_age", "distance_to_MRT", "convenience_stores",
    "latitude", "longitude", "price_per_unit"
]
```

```python
plt.figure(figsize=(16, 12))
for i, column in enumerate(df.columns):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[column], kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.tight_layout()

plt.show()
```

```python
df.columns
```

###**Bivariate Analysis**

```python
plt.figure(figsize=(10, 6))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(16, 12))
for i, column in enumerate(df.columns):
    plt.subplot(3, 3, i+1)
    sns.scatterplot(x=column, y='price_per_unit', data=df, color='purple')
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```



### Multivariate Analysis using pairplot for selected features


```python
selected_features = ["price_per_unit", "house_age", "distance_to_MRT", "convenience_stores", "latitude", "longitude"]
sns.pairplot(df[selected_features], diag_kind="kde", plot_kws={'alpha':0.6})
plt.suptitle("Multivariate Analysis - Pairplot", y=1.02)
plt.show()
```
